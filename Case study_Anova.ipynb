## Check whether all the zones generate the same amount of sales.

#### Using One Way Annova
Using One-way ANOVA :
Level of significance, alpha = 0.05
* Null Hypothesis, H0 : All zones generate the same amount of sales
* Alternate Hypothesis, H1 : All zones generate different amount of sales

```python
import scipy.stats as scp
from scipy.stats import f_oneway
```

```python
fvalue,pvalue=scp.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
fcriteria=scp.f.ppf(1-0.05,3,112)
```

```python
fcriteria=scp.f.ppf(1-0.05,3,112)
```

```python


print('F Statistics=',fvalue)
print('F criteria=',fcriteria)
print('pvalue=',pvalue)
print('significance level=0.05')

```

```python
if fvalue>fcriteria and pvalue<0.05:
   print('Reject H0: All zones generate the same amount of sales')
else:
   print('Reject H1: All zones generate different amount of sales')
```

* Sales are different is zones.
